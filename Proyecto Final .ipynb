```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv(r"C:\Users\Marcelo\Downloads\51f7ab6a-24ed-4a06-9f5d-04873f1673fd_Data-1.csv")
```

```python
df['Series Name'].unique()
```

```python
series1 = ['GDP growth (annual %)']
small_df = df[df['Series Name'].isin(series1)]
small_df.head(5)
```

```python
paises = ['Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Ecuador',  'Paraguay', 'Peru', 'Uruguay', 'Venezuela, RB']
```

```python
gdp = small_df[df['Country Name'].isin(paises)]
gdp = gdp[['Country Code','2008 [YR2008]']]
gdp['GDP Growth (annual %)']= gdp['2008 [YR2008]']
gdp = gdp[['Country Code','GDP Growth (annual %)']]
```

```python
series2 = ['Access to electricity (% of population)']
df = df[df['Series Name'].isin(series2)]
elec = df[df['Country Name'].isin(paises)]
elec = elec[['Country Code','2008 [YR2008]']]
elec['Access to electricity (% of population)']= elec['2008 [YR2008]']
elec = elec[['Country Code','Access to electricity (% of population)']]
```

```python
merge = pd.merge(gdp,
                     elec,
                    left_on = 'Country Code',
                    right_on = 'Country Code',
                    how = 'inner', indicator = True)
merge
merge['GDP Growth (annual %)'] = pd.to_numeric(merge['GDP Growth (annual %)'], errors='coerce')
merge['Access to electricity (% of population)'] = pd.to_numeric(merge['Access to electricity (% of population)'], errors='coerce')
```

```python
#GrÃ¡fico 1
alt.Chart(merge).mark_bar().encode(
    alt.Y("GDP Growth (annual %)"),
    alt.X("Country Code")
)
```

```python
#GrÃ¡fico 2
alt.Chart(merge).mark_point().encode(
    x = alt.X("GDP Growth (annual %):Q"),
    y = alt.Y("Access to electricity (% of population):Q"),
    color = "Country Code", 
    tooltip = ["Country Code", "GDP Growth (annual %)"]
    ).interactive()
```
